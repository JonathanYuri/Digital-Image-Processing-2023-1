```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

# Create the image of a paraboloid with one axis scaled (like an oval paraboloid).

```python
amount_of_points = 15
x = np.linspace(-10, 10, amount_of_points, endpoint=True)
y = np.linspace(-10, 10, amount_of_points, endpoint=True)
X, Y = np.meshgrid(x, y)

# z = x 2 + 2y 2
Z = X ** 2 + 2*Y ** 2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Oval Paraboloid')

plt.show()
```

# Create the image of a rotated sin using rotation of coordinates.

```python
amount_of_points = 100
x = np.linspace(-10, 10, amount_of_points, endpoint=True)
y = np.sin(x)

theta = np.pi / 4

# x' = x cos(theta) + y sin(theta)
x_rotated = x * np.cos(theta) + y * np.sin(theta)

# y' = -x sin(theta) + y cos(theta)
y_rotated = -x * np.sin(theta) + y * np.cos(theta)

plt.plot(x_rotated, y_rotated)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Rotated Sine Function')
plt.grid(True)
plt.show()
```

# Create the image of a gaussian.

```python
mu = 0 # mean
sigma = 1 # std deviation
amount_of_points = 100

x = np.linspace(-10, 10, amount_of_points, endpoint=True)
y = np.exp(-((x - mu) ** 2) / (2 * sigma ** 2)) / (sigma * np.sqrt(2 * np.pi))

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gaussian')
plt.grid(True)

plt.show()
```

# Create a function that generates the image of a Gaussian optionally rotated by an angle theta and with mx, my, sx, sy as input arguments.

```python
mx = float(input("Digite o valor de mx: ")) # translate in x-axis
my = float(input("Digite o valor de my: ")) # translate in y-axis
sx = float(input("Digite o valor de sx: ")) # scale in x-axis
sy = float(input("Digite o valor de sy: ")) # scale in y-axis

mu = 0 # mean
sigma = 1 # std deviation
theta = np.pi / 4 # rotation angle
amount_of_points = 100

x = np.linspace(-5, 5, amount_of_points, endpoint=True)
y = np.exp(-((x - mu) ** 2) / (2 * sigma ** 2)) / (sigma * np.sqrt(2 * np.pi))

x_rotated = x * np.cos(theta) + y * np.sin(theta) # x' = x cos(theta) + y sin(theta)
y_rotated = -x * np.sin(theta) + y * np.cos(theta) # y' = -x sin(theta) + y cos(theta)

x_transformed = (x_rotated * sx) + mx
y_transformed = (y_rotated * sy) + my

plt.plot(x_transformed, y_transformed)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gaussian With Transformations')
plt.grid(True)

plt.show()
```
